## Визуальный анализ

```python
# https://seaborn.pydata.org/examples/many_pairwise_correlations.html
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, 
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, 
            annot=True, 
            center=0, vmin=-1, vmax=1,
           )
plt.show()
```

- Стоит ли использовать все признаки?

*На тепловой карте можно отметить более сильные и менее сильные признаки которые коррелируют с таргетом.*

*Однако на данном этапе я бы не стала отбрасывать какие-либо признаки, т.к. даже слабокоррелирующие признаки могут вносить свой положительный вклад в модель и улучшить итоговое качество.*

```python
X = df.drop(['mpg','name'], axis=1)
X = sm.add_constant(X)
y = df['mpg']

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

- p_value<0.05 -> $H_0: \beta = 0$ -> отвергается - коэффициент значим, коэффициент НЕ равен 0
- p_value>0.05 -> $H_0: \beta = 0$ -> НЕ отвергается - коэффициент НЕ значим, коэффициент равен 0

*Судя по p-value можно попробовать отбросить "незначимые" признаки: cylinders, horsepower, acceleration.*

```python
X = df.drop(['mpg','name', 'cylinders', 'horsepower', 'acceleration'], axis=1)
X = sm.add_constant(X)
y = df['mpg']

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

*После удаления трех вышеперечисленных признаков R2 уменьшился - предлагается не удалять признаки.*

## Регрессия

```python
df.head(3)
```

```python
df2=pd.get_dummies(df,columns=['origin'],prefix='col', prefix_sep='',drop_first=True)
df2
```

```python
import statsmodels.formula.api as smf

model1 = smf.ols('np.log(mpg) ~ cylinders + np.log(displacement) + np.log(horsepower) +np.log(weight) + acceleration + year + col2 + col3', data=df2)
result1 = model1.fit()

print(result1.summary())
```

```python
result1.params
```

*После применения логарифма к некоторым переменным значение R2 увеличилось.*

*Признаки, которые слабее влияют на таргет - cylinders и displacement.*

```python
model2 = smf.ols('np.log(mpg) ~ cylinders + np.log(horsepower) +np.log(weight) + acceleration + year + col2 + col3', data=df2)
result2 = model2.fit()

print(result2.summary())

# удаление признака displacement не ухудшило качество модели.
```

```python
result2.params
```

```python
# Самые большие веса получились у признаков: horsepower и weight - и это логично - такие признаки действительно напрямую влияют на расход топлива.
# Признак year и преобразованный признак "origin" (col2 и col3) имеют меньшие веса, 
# однако остаются значимыми для модели. Например, year - можно объяснить тем, что новые модели авто потребяют меньше топлива, чем их предшественники.
# Самые маленькие веса имеют cylinders и acceleration, однако не смотря на показатель P-value, удаление этих признаков приводит к ухудшению модели.
```

```python
model3 = smf.ols('np.log(mpg) ~ cylinders + np.log(horsepower) +np.log(weight) + year + col2 + col3', data=df2)
result3 = model3.fit()

print(result3.summary())

# удаление признака acceleration ухудшило качество модели, данный признак не удаляем.
```

```python
# STL-декомпозиция:

plt.figure(figsize=(20,16))
sm.tsa.seasonal_decompose(df.value).plot()
plt.show()
```

### Стабилизация дисперсии

```python
from scipy import stats

df['value_boxcox'], lmbd = stats.boxcox(df['value'])
print("Оптимальный параметр преобразования Бокса-Кокса: ",lmbd)
df.head(3)
```

```python
plt.figure(figsize=(16,5))
df.value_boxcox.plot()
plt.ylabel('Transformed value')
plt.show()
# После преобразования Бокса-Кокса дисперсия ряда стала более стабильной:
```

```python
# STL-декомпозиция:

plt.figure(figsize=(20,16))
sm.tsa.seasonal_decompose(df.value_boxcox).plot()
plt.show()
```

### Дифференцирование

```python
# Сезонность:

df['value_boxcox_diff'] = df.value_boxcox - df.value_boxcox.shift(12)

plt.figure(figsize=(16,5))
df.value_boxcox_diff[12:].plot()
plt.ylabel('Differentiated and Transformed value')
plt.show()

df.head(13)
```

```python
# Сезонность пропала
```

```python
# STL-декомпозиция:

plt.figure(figsize=(20,16))
sm.tsa.seasonal_decompose(df.value_boxcox_diff[12:]).plot()
plt.show()
```

```python
# Тренд:

df['value_boxcox_diff2'] = df.value_boxcox_diff - df.value_boxcox_diff.shift(1)

plt.figure(figsize=(16,5))
df.value_boxcox_diff2[13:].plot()
plt.ylabel('2x Differentiated and Transformed value')
plt.show()
```

```python
# STL-декомпозиция:

plt.figure(figsize=(20,16))
sm.tsa.seasonal_decompose(df.value_boxcox_diff2[13:]).plot()
plt.show()
```

```python
#  Проверка стационарности ряда:

print("Критерий Дики-Фуллера для преобразованного ряда: pvalue =",
      sm.tsa.stattools.adfuller(df.value_boxcox_diff2[13:])[1])

# Гипотеза о нестационарности отвергается
```

### Остатки модели

```python
# Q-тест Льюнг — Бокса — статистический критерий, предназначенный для нахождения автокорреляции временных рядов. 
# Вместо тестирования на случайность каждого отдельного коэффициента, 
# он проверяет на отличие от нуля сразу несколько коэффициентов автокорреляции.

# H_0: данные являются случайными (то есть представляют собой белый шум).
# H_a: данные не являются случайными.

# Значение p < 0,05 H_0 отклоняется,  (остатки зависимы)
# значение p > 0,05 нулевая гипотеза не отвергается. (остатки независимы)

# Из таблицы выше мы видим, что Ljung-Box = 1.28, а это > 0.05  - значит остатки независимы.
```

```python
# Визуально остатки имеют нормальное распределение, нет смещенности или нестационарности.
```

```python
# Тест на несмещенность и стационарность: 

# Критерий знаков используется при проверке нулевой гипотезы о 
# равенстве медианы некоторому заданному значению (для одной выборки).

from statsmodels.stats.descriptivestats import sign_test
print("Критерий знаков для остатков: pvalue =",
      sign_test(best_model.resid[13:])[1]) 

# Тест Дики-Фуллера является средством проверки стационарности временного ряда.
# Нулевая гипотеза H0: g = 0 (существует единичный корень, ряд нестационарный).
# Альтернативная гипотеза: H1: g < 0 (единичного корня нет, ряд стационарный).

print("Критерий Дики-Фуллера для остатков: pvalue =",
      sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```
